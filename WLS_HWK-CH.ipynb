```r
WLS <- read.csv("WLS.csv")
summary(WLS)
```

```r
WLS[WLS == 0] <- NA
WLS[WLS == ""] <- NA
WLS[WLS < 0] <- NA
WLS$deatyr <- replace(WLS$deatyr, which(WLS$deatyr > 9000), NA)
summary(WLS)
```
